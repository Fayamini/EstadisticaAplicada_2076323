```python
#Importar librerías
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

## Inicio

```python
#Añadir los datos
personas=['Jorge','Raúl','Ana','Lissa']
edades=[27,16,21,30]

dic1={'Municipio':['Escobedo','Apodaca','Guadalupe','Santiago'],'Nacimientos':[123,345,278,321],'Muertes':[23,17,56,32],'Matrimonios':[76,45,89,12]}
datos_mun=pd.DataFrame(dic1)
```

```python
datos_mun
```

```python
personas, edades
```

```python
# Graficar
plt.bar(personas,edades)
plt.title('Edades')
plt.xlabel('Nombres')
plt.ylabel('Edad en años')
plt.show()
```

```python
plt.bar(datos_mun['Municipio'],datos_mun['Matrimonios'],color=['blue','turquoise','cyan','gray'])
plt.title('Matrimonios en N.L en 20XX')
plt.xlabel('Municipios')
plt.ylabel('Matrimonios')
plt.show()
```

```python
# Colocar más gráficas en un espacio.
# Construímos un pizarron donde colocar las gráficas
fig, (graf1,graf2)=plt.subplots(1,2)
## Gráfica 1
graf1.bar(personas,edades)
## Gráfica 2
graf2.bar(datos_mun['Municipio'],datos_mun['Matrimonios'])

plt.show()
```
