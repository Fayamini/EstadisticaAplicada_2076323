```python
## Librerías
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
pasajeros=pd.read_csv('https://raw.githubusercontent.com/jimmyzac/Estadistica-Aplicada-FCFM-UANL/main/bases_datos/airline_passengers.csv',parse_dates=True)
#Parse dates reconoce como comando de fecha 
```

```python
pasajeros
```

```python
pasajeros=pasajeros.rename(columns={'Month':'mes','Thousands of Passengers':'Miles de pasajeros'})
```

```python
plt.plot(pasajeros['mes'],pasajeros['Miles de pasajeros'])
plt.title('Miles de pasajeros de avión en Estados Unidos 1949-1960')
plt.xlabel('Fecha')
plt.ylabel('Miles de pasajeros')
plt.show()
```
